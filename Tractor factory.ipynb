```python
import numpy as np
import pandas as pd
```

```python
sales_amount = np.random.randint(30,size=(5,4))
sales_amount
```

```python
weekly_sales = pd.DataFrame(sales_amount,index=['Mon','Tue','Wed','Thu','Fri'],columns=['DongFeng','TTZ','Bel','Lovol'])
weekly_sales
```

```python
prices = np.array([150,180,200,300])
prices
```

```python
tractor_prices = pd.DataFrame(prices.reshape(1,4),index=['Prices'],columns=['DongFeng','TTZ','Bel','Lovol'])
tractor_prices
```

```python
total_prices = prices.dot(sales_amount.T)
```

```python
total_prices
```

```python
weekly_sales.T.shape
```

```python
daily_sales = tractor_prices.dot(weekly_sales.T)
```

```python
daily_sales
```

```python
weekly_sales['Total ($)'] = daily_sales.T
weekly_sales
```
