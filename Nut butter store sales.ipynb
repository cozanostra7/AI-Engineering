## Dot product example

```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(0)
sales_amount = np.random.randint(20,size=(5,3))
sales_amount
```

```python
# create weekly_sales  DataFrame
weekly_sales = pd.DataFrame(sales_amount,index=['Mon','Tue','Wed','Thu','Fri'],columns=['Almond Butter','Peanut butter','Cashew butter'])
```

```python
weekly_sales
```

```python
# Calculate prices array
prices = np.array([10,8,12])
```

```python
#create butter_prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),index=['Prices'],columns=['Almond Butter','Peanut butter','Cashew butter'])
butter_prices
```

```python
total_sales = prices.dot(sales_amount)
```

```python
# Shapes aren't aligned ,let's transpose
total_sales = prices.dot(sales_amount.T)
```

```python
total_sales
```
